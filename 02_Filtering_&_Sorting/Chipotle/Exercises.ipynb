### Step 4. How many products cost more than $10.00?

```python
# Solution 1
f_conv = lambda x: float(re.findall(r'[0-9.]+', x)[0])
chipo.insert(5, 'item_price_f', chipo['item_price'].apply(f_conv))
mask = chipo['item_price_f']>10
len(chipo.loc[mask, 'item_name'].unique())
```

```python
# Solution 2
chipo.query('item_price_f > 10').item_name.nunique()
```

```python
chipo.query('item_price_f > 10')
chipo[chipo['item_price_f']>10]
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo[mask].loc[:,['item_name', 'item_price']]
```

### Step 6. Sort by the name of the item

```python
chipo[mask].loc[:,['item_name', 'item_price']].sort_values('item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
max_price = chipo['item_price_f'].max()
print("-- price of most expensive item ordered : ", max_price)
print("\n -- most expensive item ordered: \n", chipo[chipo['item_price_f']==max_price])
print("\n -- total nb of this item ordered: ",chipo[chipo['item_price_f']==max_price]['quantity'].sum())
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo[chipo['item_name']=='Veggie Salad Bowl']['quantity'].sum()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
c1 = chipo['item_name']=='Canned Soda'
c2 = chipo['quantity']>1
ind = (chipo[c1&c2]).index
print("Canned Soda ordered twice or more : ", str(len(ind)), ' times.')
chipo.loc[ind,:]
```
